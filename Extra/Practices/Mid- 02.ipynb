```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  row, col = floor.shape
  for i in range(row):
    if i%2 == 0:
      for j in range(0, col, 2):
        print(floor[i][j], end= " ")
    else:
      for k in range(col-1, -1, -2):
        print(floor[i][k], end= " ")
    print()

floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
  row, col = m.shape
  mat = np.zeros((col, row), dtype = int)
  for i in range(row):
    for j in range(col):
      mat[j][i] = m[i][j]
  return mat


m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
import numpy as np
def arrangement(m, diff):
  row, col = m.shape
  flag = False
  for i in range(row):
    for j in range(col):
      if i != j:
        if m[i][j] - m[j][i] == diff or m[j][i] - m[i][j] == diff:
         flag = True
        else:
          return False
  return flag

m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)

m = np.array([[8,15,7,12],[13,2,18,11],[9,20,5,2],[14,9,0,10]])
print_matrix(m)
print()
returned_value = arrangement(m,2)
print(returned_value)#This should print True
unittest.output_test(returned_value, True)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
  board = np.zeros((8, 8), dtype = int)
  row, col = knight[0], knight[1]

  if 0<= row <=7 and 0<= col<= 7:
    board[row][col] = 33
    pos = np.array([(1, 2), (1, -2), (-1, 2), (-1, -2), (2, 1), (2, -1), (-2, 1), (-2, -1)])

    for i in range(len(pos)):
      if 0<= row + pos[i][0] <= 7 and 0 <= col + pos[i][1] <= 7:
        board[row+pos[i][0]][col+ pos[i][1]] = 3

  else:
    return np.zeros((0,0), dtype = int)

  return board


knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```

```python
def topple(knight,rook):
  board = np.zeros((8, 8), dtype = int)
  row, col = knight[0], knight[1]
  row2, col2 = rook[0], rook[1]
  flag = False

  if 0<= row2 <=7 and 0<= col2 <= 7:
    board[row2][col2] = 55

  if 0<= row <=7 and 0<= col<= 7:
    board[row][col] = 33
    pos = np.array([(1, 2), (1, -2), (-1, 2), (-1, -2), (2, 1), (2, -1), (-2, 1), (-2, -1)])

    for i in range(len(pos)):
      if 0<= row + pos[i][0] <= 7 and 0 <= col + pos[i][1] <= 7:
        if board[row + pos[i][0]][col + pos[i][1]] == board[row2][col2]:
          flag = True
        board[row+pos[i][0]][col+ pos[i][1]] = 3


    if flag == True:
      return (board, "Knight can kill")
    elif rook[0] == knight[0] or rook[1] == knight[1]:
        return (board, "Rook can kill")
    else:
        return (board, "Cannot kill")


knight = (3,4)
rook = (5,1)
chess_board, ans = topple(knight,rook)
print_matrix(chess_board)
print(ans) #This should print Rook can kill
```
