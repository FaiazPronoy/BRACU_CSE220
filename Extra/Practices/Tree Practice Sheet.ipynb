```python
class btnode:
  def __init__(self, elem):
    self.elem = elem
    self.left = None
    self.right = None
```

```python
def inorder(root):
  if root == None:
    return None
  else:
    inorder(root.left)
    print(root.elem, end= " ")
    inorder(root.right)
```

```python
def tree_construction(arr, i= 1):
  if i >= len(arr) or arr[i] == None:
    return None
  else:
    root = btnode(arr[i])
    root.left = tree_construction(arr, 2* i)
    root.right = tree_construction(arr, 2*i+1)
  return root


arr = [None, 1, 2, 3, 4, 5, 6]
root = tree_construction(arr)
inorder(root)
```

```python
#1
def identical(root1, root2):
  if root1 == None and root2 == None:
    return True
  elif root1 == None or root2 == None:
    return False
  elif root1 != None and root2 != None:
      if root1.elem != root2.elem:
        return False
      else:
        return identical(root1.left, root2.left) and identical(root1.right, root2.right)


root1 = tree_construction([1,2,3])
root2 = tree_construction([1,2,3])

identical(root1, root2)
```

```python
#2
def mirror_root(root):
  if root == None:
    return None
  else:
    new_root = btnode(root.elem)
    new_root.left = mirror_root(root.right)
    new_root.right = mirror_root(root.left)
  return new_root

root = tree_construction([None, 10, 20, 30, 40, 60, None, None])
inorder(root)
print()

mirr_root = mirror_root(root)
inorder(mirr_root)
```

```python
#3
def height(root):
  if root == None:
    return 0
  else:
    left_height = height(root.left)
    right_height = height(root.right)

    if left_height > right_height:
      return left_height + 1
    else:
      return right_height +1



def check_balanced_tree(root):
      left_sub = height(root.left)
      right_sub = height(root.right)

      if left_sub - right_sub > 1 or right_sub-left_sub > 1:
            return False
      else:
        return True



root = tree_construction([None, 10, 20, 30, 40, 60, None, None])
check_balanced_tree(root)

```

```python
#4
def is_parent_equal_sum_of_child(root):
  if root == None:
    return 0
  elif root != None:
     if root.left == None and root.right == None:
         return 0

     elif root.left == None and root.right != None:
        if root.right.elem == root.elem:
          return 1
        else:
          return  is_parent_equal_sum_of_child(root.right)

     elif root.right == None and root.left != None:
      if root.left.elem == root.elem:
        return 1
      else:
        return  is_parent_equal_sum_of_child(root.left)

     else:
        if root.left.elem + root.right.elem == root.elem:
          return 1
        else:
          return 0


root = tree_construction([None,7,4,3,5, None, None, None])

is_parent_equal_sum_of_child(root)
```

```python
#5
def largest_level(root, idx=0, dict1= {}):
  if root == None:
    return dict1
  elif idx not in dict1:
    dict1[idx] = root.elem
  elif idx in dict1:
    if dict1[idx] < root.elem:
      dict1[idx] = root.elem
  return largest_level(root.left, idx+1, dict1) and largest_level(root.right, idx+1, dict1)


root = tree_construction([None, 4, 9, 2, 3, 5, 7, None])
largest_level(root)
```

```python
#6
def duplicate(root, lst= []):
  if root == None:
    return False
  elif root.elem in lst:
    return True
  else:
    lst.append(root.elem)
    return duplicate(root.left, lst) or duplicate(root.right, lst)



root = tree_construction([None, 1, 2, 3, None, None, None, 4])
duplicate(root)
```

```python
#7
def print_node_of_a_level(root, k):
  if root == None:
    return None
  elif k == 0:
    print(root.elem, end= " ")
  else:
    return print_node_of_a_level(root.left, k-1) or print_node_of_a_level(root.right, k-1)



root = tree_construction([None, 1, 2, 3, 4, 5, 8, None])
print_node_of_a_level(root, 2)
```

```python
#8
def print_anchestors(root, k):
  if root == None:
    return False
  elif root.elem == k:
    return True

  else:
    if print_anchestors(root.left, k) or print_anchestors(root.right, k) == True:
      print(root.elem, end = " ")
      return True
    return False

root = tree_construction([None, 1, 2, 3, 4, 5, None, None, 7, None, None, None])
print_anchestors(root, 7)
```

```python
#9
def one_child(root, lst= []):
  if root == None:
    return None
  elif root.left == None and root.right == None:
    return None

  elif root.left == None or root.right == None:
    lst.append(root.elem)

  else:
    one_child(root.left) or one_child(root.right)
    return lst


root = tree_construction([None, 2, 3, 5, 7, None, 8, 6])
a1 = one_child(root)
if a1 != []:
  print(a1[0])
else:
  print("-1")
```

```python
#10
def skew(root):
  if root == None:
    return True
  elif root.left == None and root.right == None:
    return True
  elif root.left != None and root.right != None:
    return False
  else:
    if root.left != None:
      return skew(root.left)
    else:
      return skew(root.right)



root = tree_construction([None, 5, 4, None, None, 3, 2, None])
skew(root)
```

```python
#11
def min_depth(root):
  if root == None:
    return 0

  elif root.left == None and root.right == None:
    return 1

  elif root.left == None:
    return 1 + min_depth(root.right)

  elif root.right == None:
    return 1+ min_depth(root.left)

  else:
    return 1+ min_depth(root.left) or min_depth(root.right)




root = tree_construction([6,8, 20, None, None, 15, 7])
min_depth(root)
```

```python
def shortpath(root):
  if root==None:
    return 0
  else:
    return 1 + min(shortpath(root.left),shortpath(root.right))
root = tree_construction([6,8, 20, None, None, 15, 7])
shortpath(root)
```

```python
#12
def non_empty_child(root):
  if root == None:
    return None
  elif root.left != None and root.right != None:
    print(root.elem, end= " ")
    return non_empty_child(root.left) or non_empty_child(root.right)


root = tree_construction([None, 10, 8, 2, 3, 5, 7, None])
non_empty_child(root)
```

```python
#13
def printPath(root, k, lst=[]):
    if root == None:
        return False

    elif root.elem == k:
        lst.append(root.elem)
        return True

    else:
        lst.append(root.elem)
        if printPath(root.left,k, lst) or printPath(root.right,k, lst):
            return lst
        else:
            lst.pop()



root = tree_construction([None, 1, 2, 3, 4, 5, 6, 7])
printPath(root, 5)
```

```python
#14
def binary_expression(root):
  if root == None:
    return 0

  elif root.left == None and root.right == None:
    return root.elem

  else:
    left_sum = binary_expression(root.left)
    right_sum = binary_expression(root.right)

    if root.elem == "+":
        return left_sum + right_sum
    elif root.elem == "-":
        return left_sum - right_sum
    elif root.elem == "*":
        return left_sum * right_sum
    else:
        return left_sum // right_sum



root = tree_construction([None, "+", "*", "/", "-", 5, 21, 7, 10, 5])
binary_expression(root)
```

```python
#Creating BST/ Inserting a BST
def add_nodes(root, elem):
  if elem < root.elem and root.left == None:
    n = btnode(elem)
    root.left = n
  elif elem > root.elem and root.right == None:
    n = btnode(elem)
    root.right = n
  elif elem < root.elem and root.left != None:
    add_nodes(root.left, elem)
  elif elem > root.elem and root.right != None:
    add_nodes(root.right, elem)

arr = [70, 50, 90, 40, 60, 80, 95, 20, 75, 85, 99]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])
```

```python
#1
def get_min(root):
  temp = root
  while temp.left != None:
    temp = temp.left
  return temp.elem

def get_max(root):
  temp = root
  while temp.right != None:
    temp = temp.right
  return temp.elem


def bst_or_not(root):
  if root == None:
    return 1

  elif root.left != None and int(get_min(root)) > root.elem:
    return 0

  elif root.right != None and int(get_max(root)) < root.elem:
    return 0

  return bst_or_not(root.left) or bst_or_not(root.right)




root = tree_construction([None, 2, None, 7, None, None, None, 6])
bst_or_not(root)
```

```python
#2
def preorder(root):
  if root == None:
    return None
  else:
    print(root.elem, end= " ")
    preorder(root.left)
    preorder(root.right)



def create_balanced_bst(arr, start, end):
  if start >= end:
    return None
  else:
    mid = (start+end)//2
    root = btnode(arr[mid])
    root.left = create_balanced_bst(arr, start, mid)
    root.right = create_balanced_bst(arr, mid+1, end)
  return root

arr = [-7,4,22,44,47,51,59,76,108,132]
new_root = create_balanced_bst(arr, 0, len(arr))
preorder(new_root)
print()
inorder(new_root)

print()

```

```python
#3
def inorder_succ(root, k):
  if root == None:
    return None

  elif k < root.elem:
    left = inorder_succ(root.left, k)
    if left!= None:
      return left
    else:
      return root

  else:
    return inorder_succ(root.right, k)




arr = [20, 8, 22, 4, 12, 10, 14]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])

inorder_succ(root, 20).elem
```

```python
#3
def inorder_pre(root, k):
  if root == None:
    return None

  elif k > root.elem:
    right = inorder_pre(root.right, k)
    if right!= None:
      return right
    else:
      return root

  else:
    return inorder_pre(root.left, k)




arr = [20, 8, 22, 4, 12, 10, 14]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])
inorder(root)
inorder_pre(root, 20).elem
```

```python
#4
i = 0
def kth_largest(root, k):
    global i
    if i >= k or root==None :
        return None
    else:
      right = kth_largest(root.right, k)
      if right != None:
          return right
      else:
        i += 1

    if i == k:
        return root
    else:
      return kth_largest(root.left, k)

arr = [4,2,9]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root)
print()
k = 4
print(kth_largest(root,k).elem)
```
