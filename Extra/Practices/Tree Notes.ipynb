```python
class btnode:
  def __init__(self, elem):
    self.elem = elem
    self.left = None
    self.right = None

def tree_construction(arr, i= 1):
  if i >= len(arr) or arr[i] == None:
    return None
  else:
    root = btnode(arr[i])
    root.left = tree_construction(arr, 2*i)
    root.right = tree_construction(arr, 2*i + 1)
    return root

arr = [None, 70, 50, 90, 40, 60, 80, 95, 20, None, None, None, 75, 85, None, 99]
root = tree_construction(arr)
```

```python
def inorder(root):
  if root == None:
    return None
  else:
    inorder(root.left)
    print(root.elem, end= " ")
    inorder(root.right)
inorder(root)
```

```python
def preorder(root):
  if root == None:
    return None
  else:
    print(root.elem, end= " ")
    preorder(root.left)
    preorder(root.right)
preorder(root)
```

```python
def postorder(root):
  if root == None:
    return None
  else:
    postorder(root.left)
    postorder(root.right)
    print(root.elem, end= " ")
postorder(root)
```

```python
def get_level(root, node_level, elem):
  if root == None:
    return 0
  else:
    if root.elem == elem:
      return node_level
    elif root.left != None and elem <= root.elem:
      return get_level(root.left, node_level+1, elem)
    elif root.right != None and elem >= root.elem:
      return get_level(root.right, node_level+1, elem)

print(get_level(root, 0, 20))
```

```python
def level_depth(root):
  if root == None:
    return 0
  else:
    left = level_depth(root.left)
    right = level_depth(root.right)

    if left > right:
      return left + 1
    else:
      return right + 1

print(level_depth(root))
```

```python
def height(root):
  if root == None:
    return -1
  else:
    left_height = height(root.left)
    right_height = height(root.right)

    if left_height > right_height:
      return left_height + 1
    else:
      return right_height + 1

print(height(root))
```

```python
def no_of_nodes(root):
  if root == None:
    return 0
  else:
    return 1 + no_of_nodes(root.left) + no_of_nodes(root.right)

print(no_of_nodes(root))
```

```python
def is_tree_full(root):
  if root == None:
    return True
  elif root.left == None and root.right == None:
    return True
  elif root.left != None and root.right != None:
    return (is_tree_full(root.left) and is_tree_full(root.right))
  else:
    return False

arr = [None, 70, 50, 90, 40, 60, 80, 95, 20, 30, 55, 65, 75, 85, 99, 100]
root = tree_construction(arr)

if is_tree_full(root) == True:
  print("Full Binary Tree")
else:
  print("Not a full binary tree")
```

```python
def is_tree_complete(root, idx, no_of_nodes):
  if root == None:
    return True
  elif idx >= no_of_nodes:
    return False
  else:
    return is_tree_complete(root.left, 2*idx + 1, no_of_nodes) and is_tree_complete(root.right, 2*idx + 2, no_of_nodes)


arr = [None, 70, 50, 90, 40, 60, 80, 95, 20, 30, 55, 65, 75, 85, 99, None]
root = tree_construction(arr)

if is_tree_complete(root, 0, no_of_nodes(root)) == True:
  print("Complete Binary Tree")
else:
  print("Not a complete binary tree")
```

```python
def is_tree_perfect(root):
  if root == None:
    return True
  elif root.left == None and root.right == None:
    return True
  elif root.left == None or root.right == None:
    return False
  else:
    return is_tree_perfect(root.left) and is_tree_perfect(root.right)



arr = [None, 70, 50, 90, 40, None, 80, 95, 20, 30, 55, 65, 75, 85, 99, 100]
root = tree_construction(arr)

if is_tree_perfect(root) == True:
  print("Perfect Binary Tree")
else:
  print("Not a perfect binary tree")
```

```python
def is_identical(root1, root2):
  if root1 == None and root2 == None:
    return True

  elif root1 == None or root2 == None:
    return False

  elif root1 != None and root2 != None:
      if root1.elem != root2.elem:
           return False
      else:
         return is_identical(root1.left, root2.left) and is_identical(root1.right, root2.right)


arr = [None, 70, 50, 90, 40, 60, 80, 95, 20, None, None, None, 75, 85, None, 99]
root1 = tree_construction(arr)

arr2 = [None, 70, 50, 90, 40, 60, 80, 95, 20, None, None, None, 75, 85, None, 99]
root2 = tree_construction(arr2)


if is_identical(root1, root2) == True:
  print("Identical")
else:
  print("Not Identical")
```

```python
#Creating BST/ Inserting a BST
def add_nodes(root, elem):
  if elem < root.elem and root.left == None:
    n = btnode(elem)
    root.left = n
  elif elem > root.elem and root.right == None:
    n = btnode(elem)
    root.right = n
  elif elem < root.elem and root.left != None:
    return add_nodes(root.left, elem)
  elif elem > root.elem and root.right != None:
    return add_nodes(root.right, elem)

arr = [70, 50, 90, 40, 60, 80, 95, 20, 75, 85, 99]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])

inorder(root)
```

```python
def search(root, item):
  if root.elem == item:
    return root
  elif item < root.elem:
    if root.left == None:
      return None
    else:
      return search(root.left, item)


  elif item > root.elem:
    if root.right == None:
      return None
    else:
      return search(root.right, item)

arr = [None, 70, 50, 90, 40, 60, 80, 95, 20, 30, 55, 65, 75, 85, 99, 100]
root = tree_construction(arr)

x = search(root, 40)
if x == None:
  print("Not Found")
else:
  print("Found")
```

```python
def min_value_node(node):
  temp = node
  while temp.left != None:
    temp = temp.left
  return temp


def delete_node(root, elem):
  if root == None:
    return root
  elif elem < root.elem:
    root.left = delete_node(root.left, elem)
  elif elem > root.elem:
    root.right = delete_node(root.right, elem)
  else:
    if root.left == None:
      temp = root.right
      root = None
      return temp
    elif root.right == None:
      temp = root.left
      root = None
      return temp

    else:
      temp = min_value_node(root.right)
      root.elem = temp.elem
      root.right = delete_node(root.right, temp.elem)
  return root


arr = [70, 50, 90, 40, 60, 80, 95, 20, 75, 85, 99]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])

print("Inorder traversal of the given tree")
inorder(root)


print("\nDelete 20")
root = delete_node(root, 20)
print("Inorder traversal of the modified tree")
inorder(root)


print("\nDelete 100")
root = delete_node(root, 100)
print("Inorder traversal of the modified tree")
inorder(root)


print("\nDelete 70")
root = delete_node(root, 70)
print("Inorder traversal of the modified tree")
inorder(root)
```

```python
def haspathsum(root, target):
  if root == None:
    return False
  elif target == root.elem:
    return True
  else:
    return haspathsum(root.left, target-root.elem) or haspathsum(root.right, target-root.elem)


root = tree_construction([None, 5, 4, 8, 11, None, 13, 17, 7, 2, None, 1, None, None])
haspathsum(root, 22)
```

```python
def hailstone(n):
  if n == 0:
    return None
  elif n == 1:
    return "1"
  elif n %2 == 0:
    return str(n)+ ", " + hailstone(n//2)
  elif n %2 != 0:
    return str(n)+ ", " + hailstone(3*n+1)

hailstone(7)
```
