```python

def sumDigits(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return sumDigits(n//10) + (n % 10)

sumDigits(126)
```

```python
def strcount(str, sub):
  if str == "":
    return 0
  elif len(str) < len(sub):
    return 0
  else:
    n = len(sub)
    if str[:n] == sub:
      return 1 + strcount(str[n::], sub)
    else:
      return strcount(str[1::], sub)

strcount("catcowcat", "cat")
```

```python
def nestParen(str):
  if len(str)==0:
    return True
  else:
    n = len(str)
    if str[0] == "(" and str[n-1] == ")":
      return nestParen(str[1:n-1])
    else:
      return False

nestParen("(())")
```

```python
def strdist(str, sub):
  if str == "":
    return 0
  elif len(str) < len(sub):
    return 0

  else:
    n = len(sub)
    if str[:n] != sub and str[-n:] != sub:
      return strdist(str[1:-1], sub)
    else:
      return len(str)


strdist("catcowcat", "cow")
```

```python
def groupSum(idx, arr, sum):
    if idx == len(arr):
        if sum == 0:
          return True
        else:
          return False

    else:
        return groupSum(idx + 1, arr, sum - arr[idx]) or groupSum(idx + 1, arr, sum)



# Test cases
print(groupSum(0, [2, 4, 8], 10))   # Output: True
print(groupSum(0, [2, 4, 8], 14))   # Output: True
print(groupSum(0, [2, 4, 8], 9))    # Output: False

```

```python
def recursive_helper(arr, idx, sum1, sum2):
      if idx == len(arr):
           if sum1 == sum2:
            return True
           else:
            return False

      else:
           return (recursive_helper(arr, idx + 1, sum1 + arr[idx], sum2) or
                recursive_helper(arr, idx + 1, sum1, sum2 + arr[idx]))


def splitArray(arr):
    return recursive_helper(arr, 0, 0, 0)




splitArray([3, 2, 5])
```

```python
def splitOdd10(idx, arr, sum1, sum2):
    if idx >= len(arr):
        if (sum1 % 10 == 0 and sum2 % 2 != 0) or (sum2 % 10 == 0 and sum1 % 2 != 0):
          return True
        else:
          return False
    else:
          return splitOdd10(idx + 1, arr, sum1 + arr[idx], sum2) or splitOdd10(idx + 1, arr, sum1, sum2 + arr[idx])

def splitOdd10_possible(arr):
    return splitOdd10(0, arr, 0, 0)


print(splitOdd10_possible([5, 5, 5]))
print(splitOdd10_possible([5, 5, 6]))
print(splitOdd10_possible([5, 5, 6, 1]))
```

```python
def split53(idx, arr, sum1, sum2):
    if idx >= len(arr):
        if sum1 == sum2:
          return True
        else:
          return False

    if arr[idx] % 5 == 0:
        return split53(idx + 1, arr, sum1 + arr[idx], sum2)
    elif arr[idx] % 3 == 0:
        return split53(idx + 1, arr, sum1, sum2 + arr[idx])
    else:
        return split53(idx + 1, arr, sum1 + arr[idx], sum2) or split53(idx + 1, arr, sum1, sum2 + arr[idx])


def split53_possible(arr):
    return split53(0, arr, 0, 0)


print(split53_possible([1, 1]))
print(split53_possible([1, 1, 1]))
print(split53_possible([2, 4, 2]))
```

```python
def groupSum5(idx, arr, sum):
    if idx == len(arr):
        if sum == 0:
          return True
        else:
          return False

    elif arr[idx] % 5 == 0:
        return groupSum5(idx + 1, arr, sum - arr[idx])

    else:
        return groupSum5(idx + 1, arr, sum - arr[idx]) or groupSum5(idx + 1, arr, sum)




print(groupSum5(0, [2, 5, 10, 4], 19))
print(groupSum5(0, [2, 5, 10, 4], 17))
print(groupSum5(0, [2, 5, 10, 4], 12))
```
