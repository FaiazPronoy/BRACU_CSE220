```python
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

```python
def zigzag_walk(floor):
  row, col = floor.shape
  for i in range(0, row):
    if i % 2==0:
      for j in range(0, col, 2):
        print(floor[i][j], end= " ")
    else:
      for k in range(col-1, 0,  -2):
        print(floor[i][k], end= " ")
    print()



floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
zigzag_walk(floor)
```

```python
def landscape(mat):
  row, col = mat.shape
  tran=np.zeros((col,row),dtype=int)
  for i in range(row):
    for j in range(col):
      tran[j][i] = mat[i][j]
  return tran


mat = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(mat)
print()
print_matrix(landscape(mat))
```

```python
def arrangement(m, num):
  row, col = m.shape
  for i in range(row):
    for j in range(col):
      if m[i][j] - m[j][i] == num:
        return True
      else:
        return False





m = np.array([[8, 15, 7, 12], [13, 2, 18, 11], [9, 20, 5, 2], [14, 9, 0, 10]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)
```
