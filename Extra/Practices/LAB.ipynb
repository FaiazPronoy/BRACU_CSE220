```python
def resize(old, size):
  new = [None] * size
  for i in range(len(old)):
    new[i] = old[i]
  return new

array = [1, 2, 3, 4, 5]
size = 8
print(resize(array, size))
```

```python
arr = [10, 20, 30 ,40,50]

def leftshift(source):
  for item in range(0, len(source)-1, 1):
    source[item] = source[item+1]
  source[len(source)-1] = None
  return source
print(leftshift(arr))
```

```python
arr = [10, 20, 30 ,40,50]

def rightshift(source):
  for item in range(len(source)-1, 0,  -1):
    source[item] = source[item-1]
  source[0] = None
  return source
print(rightshift(arr))
```

```python
arr=[10,20,30,40,50,60]

def rotate_left(arr):
  temp = arr[0]
  for item in range(1, len(arr)):
    arr[item - 1] = arr[item]
  arr[len(arr)-1] = temp
  return arr

print(rotate_left(arr))
```

```python
arr=[10,20,30,40,50,60]

def rotate_right(arr):
  temp = arr[len(arr)-1]
  for item in range(len(arr)-1, 0, -1):
    arr[item] = arr[item-1]
  arr[0] = temp
  
  return arr

print(rotate_right(arr))
```
