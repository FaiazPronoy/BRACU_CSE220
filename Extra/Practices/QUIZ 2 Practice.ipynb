```python
import numpy as np

class Node:
  def __init__(self, elem, next):
    self.elem = elem
    self.next = next

def createlist(arr):
    head = Node(arr[0], None)
    tail = head
    for i in range(1, len(arr)):
      newNode = Node(arr[i], None)
      tail.next = newNode
      tail = tail.next
    return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = "-->")
    else:
      print(temp.elem)
    temp = temp.next





head = createlist(np.array([10,15,34,41,56,72]))
printLinkedList(head)
```

```python
class Queue:
  def __init__(self):
    self.front = None
    self.back = None
  def enqueue(self, elem):
    if self.front == None:
      self.front = Node(elem, None)
      self.back = self.front
    else:
      newNode = Node(elem, None)
      self.back.next = newNode
      self.back = self.back.next


  def dequeue(self):
    if self.front == None:
      print("Queue Underflow")
    else:
      deq_item = self.front.elem
      self.front = self.front.next
      return deq_item

  def peek(self):
    if self.front ==None:
      print("Queue underflow exception")
    else:
      return self.front.elem

Queue = Queue()
Queue.enqueue(2)
Queue.enqueue(3)
Queue.enqueue(4)
Queue.peek()
Queue.dequeue()
Queue.peek()
Queue.dequeue()
Queue.peek()
```
