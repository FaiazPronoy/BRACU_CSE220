```python
class btnode:
  def __init__(self, elem):
    self.elem = elem
    self.left = None
    self.right = None
def inorder(root):
  if root == None:
    return None
  else:
    inorder(root.left)
    print(root.elem, end= " ")
    inorder(root.right)
def tree_construction(arr, i= 1):
  if i >= len(arr) or arr[i] == None:
    return None
  else:
    root = btnode(arr[i])
    root.left = tree_construction(arr, 2*i)
    root.right = tree_construction(arr, 2*i + 1)
    return root

arr = [None, 70, 50, 90, 40, 60, 80, 95, 20, None, None, None, 75, 85, None, 99]
```

```python
def identical(root1, root2):
  if root1 == None and root2 == None:
    return True
  elif root1 == None or root2 == None:
    return False
  else:
    if root1.elem != root2.elem:
      return False
    return identical(root1.left, root2.left) and identical(root1.right, root2.right)





arr = [None,1,2,3]
root1 = tree_construction(arr)
arr2 = [None,1,2]
root2 = tree_construction(arr2)
identical(root1, root2)
```

```python
def mirror(root):
  if root == None:
    return None
  else:
    new_root = btnode(root.elem)
    new_root.left = mirror(root.right)
    new_root.right = mirror(root.left)
  return new_root


arr = [None,1,2,3]
root = tree_construction(arr)
inorder(root)
new = mirror(root)
inorder(new)
```

```python
def height(root):
  if root == None:
    return 0
  else:
    left = height(root.left)
    right = height(root.right)

    if left > right:
      return left +1
    else:
      return right+1






def balanced_or_not(root):
  left = height(root.left)
  right = height(root.right)

  if left - right > 1 or right - left >1:
    return False
  else:
    return True





arr = [None,10,20,30,40,60,None,None]
root = tree_construction(arr)

balanced_or_not(root)
```

```python

def is_parent_equal_sum_of_child(root):
  if root == None:
    return 0
  elif root != None:
    if root.left == None and root.right == None:
      return 0
    elif root.left == None and root.right != None:
      if root.right.elem == root.elem:
        return 1
      else:
        return is_parent_equal_sum_of_child(root.right)

    elif root.right == None and root.left != None:
      if root.left.elem == root.elem:
        return 1
      else:
        return is_parent_equal_sum_of_child(root.left)

    else:
      if root.left.elem + root.right.elem == root.elem:
        return 1
      else:
        return 0





root = tree_construction([None,7,4,3,5, None, None, None])

is_parent_equal_sum_of_child(root)
```

```python
def largest_level(root, idx = 0, dict1= {}):
  if root == None:
    return dict1
  elif idx not in dict1:
    dict1[idx] = root.elem
  elif idx in dict1:
    if dict1[idx] < root.elem:
      dict1[idx] == root.elem
  return largest_level(root.left, idx+1, dict1) and largest_level(root.right, idx+1, dict1)







root = tree_construction([None, 4, 9, 2, 3, 5, 7, None])
largest_level(root)
```

```python
def no_of_nodes(root):
  if root == None:
    return 0
  else:
    return 1 + no_of_nodes(root.left) + no_of_nodes(root.right)



root = tree_construction([None, 4, 9, 2, 3, 5, 7, None])
no_of_nodes(root)
```

```python
def add_nodes(root, elem):
  if root == None:
    return None
  elif elem < root.elem and root.left == None:
    n = btnode(elem)
    root.left = n
  elif elem > root.elem and root.right == None:
    n = btnode(elem)
    root.right = n
  elif elem < root.elem and root.left != None:
    return add_nodes(root.left, elem)
  elif elem > root.elem and root.right != None:
    return add_nodes(root.right, elem)



arr = [15,18,28,-1,16,3,6,23,-2,-6]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])

inorder(root)
```

```python
def inorder_suc(root, k):
  if root == None:
    return None
  elif k < root.elem:
    left = inorder_suc(root.left, k)
    if left != None:
      return left
    else:
      return root
  else:
        return inorder_suc(root.right, k)


arr = [20, 8, 22, 4, 12, 10, 14]
root = btnode(arr[0])

for i in range(1, len(arr)):
  add_nodes(root, arr[i])
inorder(root)
print()
inorder_suc(root, 8).elem
```

```python
i = 0
def kth_smallest(root, k):
    global i
    if i >= k or root==None :
        return None
    else:
      left = kth_smallest(root.left, k)
      if left != None:
          return left
      else:
        i += 1

    if i == k:
        return root
    else:
      return kth_smallest(root.right, k)

#DRIVER CODE
root = tree_construction([None, 9,5,11,4,7,None,12,None,None,6,8,None,None,None,None])
print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root)
print()
k = 4
print(kth_smallest(root,k).elem)
```

```python
def helper(root):
    sum = 0
    if root is None:
        return 0
    elif root.left is None and root.right is None:
        sum += root.elem
    elif root.left != None and root.right != None:
      sum += root.elem + root.left.elem + root.right.elem
    return sum


def sum_left_right(root):
    left = helper(root.left)
    right = helper(root.right)
    total = left - right
    return total

root = tree_construction([None, 10, 8, 2, 3, 5, 7, 2])
result = sum_left_right(root)
print(result)
```
