```python
def binary_search(arr, left, right, key):
    mid = left + (right-left)//2
    if left > right:
      return False
    elif arr[mid] == key:
      return True
    elif arr[mid] > key:
      return binary_search(arr, left, mid-1, key)
    else:
      return binary_search(arr, mid+1, right, key)


arr= [20, 10, 50, 40, 30, 70, 80]

print(binary_search(arr, 0, len(arr)-1, 10))
```

```python
def max_elem(arr, left, right):

    if left == right:
      return arr[left]


    mid = left + (right-left)//2
    max_left = max_elem(arr, left, mid)
    max_right = max_elem(arr, mid+1, right)

    if max_left > max_right:
      return max_left
    else:
      return max_right



arr= [20, 10, 50, 40, 30, 5, 10]

print(max_elem(arr, 0, len(arr)-1))
```
