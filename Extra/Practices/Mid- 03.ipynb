```python
#Run this cell
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  temp = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    temp.next = newNode
    temp = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

Assemble Conga Line

```python
def conga_line(head):
  temp = head
  flag = False
  while temp.next != None:
    if temp.elem < temp.next.elem:
      flag = True
    else:
      return False
    temp = temp.next
  return flag



head = createList(np.array([10,15,34,41,56,72]))
print('Original Conga Line: ', end = ' ')
printLinkedList(head)
returned_value = conga_line(head)
print(returned_value) #This should print True
unittest.output_test(returned_value, True)
print()

head = createList(np.array([10,15,44,41,56,72]))
print('Original Conga Line: ', end = ' ')
printLinkedList(head)
returned_value = conga_line(head)
print(returned_value) #This should print False
unittest.output_test(returned_value, False)
print()


```

Shuffle on index

```python
def shuffle_on_index(head):
    even_head = None
    even_temp = None
    odd_head = None
    odd_temp = None
    index = 0
    temp = head

    while temp != None:
        if index % 2 == 0:
            if even_head == None:
                even_head = temp
                even_temp = temp
            else:
                even_temp.next = temp
                even_temp = temp
        else:
            if odd_head == None:
                odd_head = temp
                odd_temp = temp
            else:
                odd_temp.next = temp
                odd_temp = temp

        index += 1
        temp = temp.next

    if even_temp != None:
        even_temp.next = odd_head

    if odd_temp != None:
        odd_temp.next = None

    return even_head


head = createList(np.array(['S','E','N','P','A','I']))
print('Original Play List: ', end = ' ')
printLinkedList(head)
newhead =  shuffle_on_index(head)
print('New Play List: ', end = ' ')
printLinkedList(newhead) #This should print S-->N-->A-->E-->P-->I
print()

head = createList(np.array(['N','I','S','H','I','N','O','Y','A']))
print('Original Play List: ', end = ' ')
printLinkedList(head)
newhead =  shuffle_on_index(head)
print('New Play List: ', end = ' ')
printLinkedList(newhead) #This should print N-->S-->I-->O-->A-->I-->H-->N-->Y
print()
```

```python
def find_next_clue(head):
  temp = head

  while temp != None:
    tail = temp
    max = 0

    while tail != None:
      if tail.elem > max:
        max = tail.elem
      tail = tail.next

    if max <= temp.elem:
      print(f'No next element greater than {max}')
    elif max > temp.elem:
      print(f'{max} is greater than {temp.elem}')
    temp = temp.next

head = createList(np.array([7,85,54,16,11,30]))
print('Given Clue List: ', end = ' ')
printLinkedList(head)
find_next_clue(head)
print()

head = createList(np.array([20,13,33,12]))
print('Given Clue List: ', end = ' ')
printLinkedList(head)
find_next_clue(head)
print()
```

```python
class Node:
  def __init__(self, element, next=None):
    self.element = element
    self.next = next

def createLinkedList(arr):
  head = Node(arr[0])
  tail = head
  for idx in range(1, len(arr)):
    newNode = Node(arr[idx])
    tail.next = newNode
    tail = newNode
  return head

def deleteLastElement(head, value):
    temp = head
    prev = None
    last_occur = None

    while temp != None:
        if temp.element == value:
            last_occur = prev
        prev = temp
        temp = temp.next

    if last_occur != None:
        last_occur.next = last_occur.next.next

    elif head.element == value:
        head = head.next

    return head

def print_list(head):
    temp = head
    while temp != None:
        if temp.next != None:
            print(temp.element, end=' ')
        else:
            print(temp.element)
        temp = temp.next
    print()

head = createLinkedList([1, 2, 3, 5, 2, 10])
new_head = deleteLastElement(head, 2)
print_list(new_head)
```
