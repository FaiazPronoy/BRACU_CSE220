```python
arr = [10, 20, 30, 40, 50]
for item in range(len(arr)):
  print(arr[item], end= " ")
```

```python
arr = [10, 20, 30, 40, 50]
for item in range(len(arr)-1, -1, -1):
  print(arr[item], end= " ")
```

```python
arr = [10, 20 , 30 , 40 ,50]

def copy_array(source):
  new_array = [None]* len(source)
  for item in range(0, len(source), 1):
    new_array[item] = source[item]
  return new_array
  
print(copy_array(arr))
```
