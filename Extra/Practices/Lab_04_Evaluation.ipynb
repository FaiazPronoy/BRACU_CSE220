```python
class Node:
  def __init__(self, element, next=None):
    self.element = element
    self.next = next

def createLinkedList(arr):
  head = Node(arr[0])
  tail = head
  for idx in range(1, len(arr)):
    newNode = Node(arr[idx])
    tail.next = newNode
    tail = newNode
  return head

def deleteLastElement(head, value):
    temp = head
    prev = None
    last_occur = None

    while temp != None:
        if temp.element == value:
            last_occur = prev
        prev = temp
        temp = temp.next

    if last_occur != None:
        last_occur.next = last_occur.next.next

    elif head.element == value:
        head = head.next

    return head

def print_list(head):
    temp = head
    while temp != None:
        if temp.next != None:
            print(temp.element, end=' ')
        else:
            print(temp.element)
        temp = temp.next
    print()

head = createLinkedList([1, 2, 3, 5, 2, 10])
new_head = deleteLastElement(head, 2)
print_list(new_head)
```
