```python
class Node:
  def __init__(self,elem):
    self.elem = elem
    self. left = None
    self. right = None
```

```python
def inorder_Traversal(root, in_order):
  if root == None:
    return 0
  else:
    inorder_Traversal(root.left, in_order)
    in_order.append(root.elem)
    inorder_Traversal(root.right, in_order)

def find_Path(root, key, lst1= []):
    if root.elem == key:
      lst1.append(root.elem)
      return lst1

    elif root.elem > key and root.left == None:
      lst1.append(root.elem)

    elif root.elem < key and root.right == None:
      lst1.append(root.elem)

    elif root.elem > key and root.left != None:
      lst1.append(root.elem)

      return find_Path(root.left, key, lst1)

    elif root.elem < key and root.right != None:

      lst1.append(root.elem)
      return find_Path(root.right, key, lst1)




```

```python
root = Node(10)
root.left = Node(30)
root.right = Node(7)
root.left.left = Node(8)
root.left.right = Node(11)
root.right.left = Node(12)
root.right.right = Node(20)
in_order_Tree=[]
inorder_Traversal(root,in_order_Tree)
print("Inorder Traversal")
print(in_order_Tree)
print("Path")
print(find_Path(root,20))
print(find_Path(root,50))
```
