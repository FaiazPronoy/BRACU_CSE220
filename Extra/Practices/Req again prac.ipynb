```python
def sumDigits(n):
  if n == 1:
    return 1
  else:
    return sumDigits(n//10) + n%10

sumDigits(126)
```

```python
def bunnyEars2(n):
  if n == 0:
    return 0
  else:
    if n % 2 == 0:
     return 3 + bunnyEars2(n-1)
    else:
      return 2 + bunnyEars2(n-1)


bunnyEars2(3)
```
