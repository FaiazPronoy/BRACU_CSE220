```python
class Node:
  def __init__(self,elem):
    self.elem = elem
    self. left = None
    self. right = None
```

```python
def inorder_Traversal(root,in_order):
  if root == None:
    return None
  else:
    inorder_Traversal(root.left, in_order)
    in_order.append(root.elem)
    inorder_Traversal(root.right, in_order)


def find_path(root,key, lst= []):
  if root == None:
    return None
  elif root.elem == key:
    lst.append(root.elem)
    return lst
  else:
    if key < root.elem and root.left != None:
      lst.append(root.elem)
      return find_path(root.left, key, lst)
    elif key > root.elem and root.right != None:
      lst.append(root.elem)
      return find_path(root.right, key, lst)
  return False

```

```python
def check_palindrome(word):
  if len(word) == 0:
    return True
  else:
    if word[0] == word[len(word)-1]:
      return check_palindrome(word[1:len(word)-1])
    else:
      return False







words = ['deified', 'nested', 'racecar', 'rabbit', 'glossy', 'level', 'discrete', 'kayak', 'stats']
print([check_palindrome(word) for word in words])
```

```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None

def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)

def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None

  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def find_nodes_at_distance_k(root, k, nodes_at_k):
  if root == None:
    return
  elif k == 0:
    nodes_at_k.append(root.val)
  else:
   return find_nodes_at_distance_k(root.left, k-1, nodes_at_k) or find_nodes_at_distance_k(root.right, k-1, nodes_at_k)

def hash_function(elem, size):
  sum = 0
  for i in range(len(elem)):
    sum += ord(elem[i])
  key = sum*len(elem) % size
  return key



def ht_insert(value, ht):
  key = hash_function(value, 4)
  if ht[key] == None:
    ht[key] = value

  else:
        for i in range(1, len(ht)):
            new_key = (key + i) % len(ht)
            if ht[new_key] == None:
                ht[new_key] = value
                break

def create_hash_table(node_at_k, h_table):
  for i in node_at_k:
    ht_insert(i, h_table)
  return h_table
```
