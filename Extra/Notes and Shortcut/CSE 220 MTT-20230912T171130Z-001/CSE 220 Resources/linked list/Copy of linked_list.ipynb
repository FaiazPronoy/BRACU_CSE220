```python
class Node:
  def __init__(self, element, next_node):
    self.element = element
    self.next = next_node
```

```python
arr = [10, 20, 30, 40, 50, 60]
def create_list(arr):
  head = Node(arr[0], None)
  cur = head
  for i in range(1, len(arr)):
    node = Node(arr[i], None)
    cur.next = node
    cur = node

  return head
```

```python
def iterate(head):
  cur = head
  while cur!=None:
    print(cur.element, end=" ")
    cur = cur.next
```

```python
head = create_list(arr)
iterate(head)
```

```python
def count_nodes(head):
  cur = head
  count = 0
  while cur!=None:
    count += 1
    cur = cur.next
  return count

print(count_nodes(head))
```

```python
def element_at(head, idx):
  cur = head
  cur_idx = 0

  if count_nodes(head) <= idx:
    return "invalid index"

  while cur != None:
    if cur_idx == idx:
      return cur.element
    else:
      cur_idx += 1
      cur = cur.next

print(element_at(head, 3))
```

```python
def node_at(head, idx):
  cur = head
  cur_idx = 0

  if count_nodes(head) <= idx:
    return None

  while cur != None:
    if cur_idx == idx:
      return cur
    else:
      cur_idx += 1
      cur = cur.next

print(node_at(head, 3))
print(node_at(head, 3).element)
```

```python
def insert(head, elem, idx):
  total_nodes = count_nodes(head)
  if idx==0:
    new_head = Node(elem, head)
    head = new_head

  elif idx>0 and idx<total_nodes:
    new_node = Node(elem, None)
    n1 = node_at(head, idx-1)
    n2 = node_at(head, idx)
    n1.next = new_node
    new_node.next = n2

  elif idx==total_nodes:
    new_node = Node(elem, None)
    n1 = node_at(head, total_nodes-1)
    n1.next = new_node

  return head


head = insert(head, 5, 0)
head = insert(head, 15, 4)
head = insert(head, 25, 8)
iterate(head)
```

```python
def copy_list(source_head):
  if source_head == None:
    return None

  copy_head = Node(source_head.element, None)
  cur = source_head.next
  copy_cur = copy_head

  while cur != None:
    copy_node = Node(cur.element, None)
    copy_cur.next = copy_node
    copy_cur = copy_node
    cur = cur.next

  return copy_head


copy_head =  copy_list(head)
iterate(copy_head)
```

```python
head  = create_list(arr)
def reverse_in_place(head):
  prev = None
  cur = head
  # nxt = cur.next

  while cur!=None:
    nxt = cur.next
    cur.next = prev
    prev = cur
    cur = nxt

  return prev

iterate(head)
print()
print('--------------------')
rev_head = reverse_in_place(head)
iterate(rev_head)
```
