```python
# A = row1 * col1 
# B = row2 * col2 

def matrix_multiplication(A, B, row1, col1, row2, col2):
  result = np.zeros(row1, col2)

  if col1 != row2:
    return "Invalid Dimensions"

  for i in range(row1):
    for j in range(col2):
      for k in range(col1): #row2
        result[i][j] += A[i][k] * B[k][j]

  return result
```

```python
# A = row1 * col1 
# B = row2 * col2 
import numpy as np
def matrix_multiplication(A, B):
  row1 = len(A)
  col1 = len(A[0])
  row2 = len(B)
  col2 = len(B[0])

  if col1 != row2:
    return "invalid dimensions"

  C = np.zeros((row1, col2))

  for i in range(row1):
    for j in range(col2):
      #####c[i][j] = row i . col j####
      for k in range(col1): #row2
        C[i][j] += A[i][k] * B[k][j]

  return C
```
