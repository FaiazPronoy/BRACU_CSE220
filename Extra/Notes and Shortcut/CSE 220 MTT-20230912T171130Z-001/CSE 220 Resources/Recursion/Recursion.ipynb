```python
def factorial(n):
  if n==1:
    return 1

  return n*factorial(n-1)
```

```python
factorial(5)
```

```python
def exponent(a,n):
  if n==1:
    return a

  return a*exponent(a,n-1)
```

```python
exponent(2,7)
import time

# ts stores the time in seconds
ts = time.time()
print(exponent(2,1002))
te = time.time()
print(te-ts)
```

```python
def fib(n):
  if n==0 or n==1:
    return n

  return fib(n-1) + fib(n-2)
```

```python
print(fib(2))
print(fib(3))
print(fib(4))
print(fib(5))
print(fib(6))
```

```python
def sum_of(arr,n):
  if n==0:
    return arr[0]

  return arr[n] + sum_of(arr,n-1)
```

```python
arr=[3, 5, 2, 6, 8]
print(sum_of(arr,4))
```

```python
def find_key(arr, n , key):
  if arr[n] == key:
    return True

  if n==-1:
    return False

  return find_key(arr, n-1, key)
```

```python
arr=[3, 5, 2, 6, 8]
print(find_key(arr, 4, 2))
print(find_key(arr, 4, 9))
```

```python
def find_key_v2(arr, n , key):
  if arr[n] == key:
    return n

  if n==-1:
    return -1

  return find_key_v2(arr, n-1, key)
```

```python
arr=[3, 5, 2, 6, 8]
print(find_key_v2(arr, 4, 2))
print(find_key_v2(arr, 4, 9))
```

```python
def efficient_expo(a,n):
  if n==1:
    return a

  if n%2 == 0:
    return efficient_expo(a,n//2) * efficient_expo(a,n//2)

  return efficient_expo(a,n//2) * efficient_expo(a,n//2) * a
```

```python
import time

# ts stores the time in seconds
ts = time.time()
print(efficient_expo(2,1000))
te = time.time()
print(te-ts)
```

```python
def binary_search(arr, l , r , key):

  # l = left most index of the arr
  # r = right most index of the arr
  # key = element to be found
  # mid = (l+r)//2  ### mid = (l+r)//2
  mid = l + (r-l)//2

  print(l, r, mid, arr[mid])

  if l > r:
    return False, -1

  if arr[mid] == key:
    return True, mid

  if arr[mid] > key:
    return binary_search(arr, l, mid-1, key)

  return binary_search(arr, mid+1, r, key)
```

```python
def max_in(arr, l , r):

  # l = left most index of the arr
  # r = right most index of the arr
  # key = element to be found
  # mid = (l+r)//2  ### mid = (l+r)//2
  if l==r:
    return arr[l]

  mid = l + (r-l)//2

  max_of_left = max_in(arr, l, mid)

  max_of_right = max_in(arr, mid+1, r)

  if max_of_left > max_of_right:
    return max_of_left

  return max_of_right
```

```python
arr=[3, 5,9, 2, 6, 8]
print(max_in(arr, 0, len(arr)-1))
```

```python
def is_palindrome(str, l, r):
  if str[l]!=str[r]
    return False

  return is_palindrome(str, l+1, r-1)
```
