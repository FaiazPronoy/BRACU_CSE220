```python
def factorial(n):
  if n==0 or n==1:
    return 1

  return n*factorial(n-1)
```

```python
factorial(5)
```

```python
def exponent(a,n):
  if n==0:
    return 1

  if n==1:
    return a

  return a*exponent(a,n-1)
```

```python
import time

# ts stores the time in seconds
ts = time.time()
print(exponent(2,1000))
te = time.time()
print(te-ts)
```

```python
def fibonacci(n):
  if n==0 or n==1:
    return n

  return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(7)
```

```python
def sum_of(arr, n):
  if n==0:
    return 0

  return arr[n-1] + sum_of(arr,n-1)
```

```python
arr = [2, 10 , 9, 5, 6]
sum_of(arr, 5)
```

```python
# n = size
def find_key(arr, n , key):

  if arr[n-1] == key:
    return True

  if n==0:
    return False

  return find_key(arr, n-1 , key)
```

```python
arr=[3, 5, 2, 6, 8]
print(find_key(arr, 5, 2))
print(find_key(arr, 5, 9))
```

```python
# n = size
def find_key_index(arr, n , key):

  if arr[n-1] == key:
    return n-1

  if n==0:
    return -1

  return find_key_index(arr, n-1 , key)
```

```python
arr=[3, 5, 2, 6, 8]
print(find_key_index(arr, 5, 2))
print(find_key_index(arr, 5, 9))
```

```python
def find_max(arr, last_index):

  if last_index==0:
    return arr[0]

  smaller_array_max = find_max(arr, last_index-1)
  if smaller_array_max > arr[last_index]:
    return smaller_array_max

  return arr[last_index]
```

```python
arr=[3, 5, 2, 6, 8]
print(find_max(arr, 4))
```

```python
def efficient_expo(a,n):
  if n==1:
    return a

  if n%2 == 0:
    return efficient_expo(a,n//2) * efficient_expo(a,n//2)

  return efficient_expo(a,n//2) * efficient_expo(a,n//2) * a
```

```python
import time

# ts stores the time in seconds
ts = time.time()
print(efficient_expo(2,1000))
te = time.time()
print(te-ts)
```
