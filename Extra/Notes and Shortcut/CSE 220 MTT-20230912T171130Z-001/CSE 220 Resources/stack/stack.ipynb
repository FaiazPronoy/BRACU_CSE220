```python
class Node:
  def __init__(self, element, next):
    self.element = element
    self.next = next
```

```python
class Stack:
  def __init__(self):
    self.top = None  #self.top = Node(None, None)

  def push(self, val):
    new_node = Node(val, None)
    new_node.next = self.top
    self.top = new_node

  def pop(self):  #stack empty
    if self.top == None:
      return None

    ele = self.top.element
    self.top = self.top.next
    return ele


  def peek(self):
    if self.top == None:
      return None

    return self.top.element


  def is_empty(self):
    return (self.top==None)



```

```python
stack = Stack()
stack.push(3)
stack.push(5)
stack.push(7)
stack.push(9)
```

```python
def rev_string(str):
  new_stack = Stack()

  for i in range(len(str)):
    new_stack.push(str[i])

  rev_str = ""

  #while new_stack.is_empty()==False:
  while not new_stack.is_empty():
    rev_str+=new_stack.pop()

  return rev_str
```

```python
print(rev_string("DataStructures"))
```

```python
def evaluate(var1, var2, operator):
  #var1 = second top
  #var2 = top

  if operator=="+":
    return var1+var2

  if operator=="-":
    return var1-var2

  if operator=="*":
    return var1*var2

  if operator=="/":
    return var1/var2
```

```python
def evaluate_exp(str):
  operators = ['+', "-", "*", "/"]
  stack = Stack()

  for i in range(len(str)):
    if str[i] in operators:
      var2 = stack.pop()
      var1 = stack.pop()
      stack.push(evaluate(int(var1), int(var2), str[i]))

    else: #if its a number
      stack.push(str[i])


  return stack.peek()

```

```python
evaluate_exp("35+2/")
```

```python
def parenthesis_balance(exp):
  opening = [ '(', '{', '[' ]
  closing = [')', '}', ']']

  opening_of = {')':'(',
                '}':'{',
                ']':'['
                }

  # opening_of[')'] = '('

  stack = Stack()

  for x in exp:
    if x in opening:
      stack.push(x)

    elif x in closing:
      if stack.is_empty():
        return "Closing bracket was found, but not opening"

      else:
        y = stack.pop()

        if opening_of[x] != y:
          return "Opening and Closing brackets do not match"

  if stack.is_empty() == True:
    return True

  else:
    return "Opening bracket(s) were still there after parsing the expression"

  # return stack.is_empty()



```

```python
parenthesis_balance("(a+b)")
```

```python
parenthesis_balance("[(a+b)")
```

```python
parenthesis_balance("(a+b)]")
```
