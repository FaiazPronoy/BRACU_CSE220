```python
class Node:
  def __init__(self, element, next):
    self.element = element
    self.next = next
```

```python
class Stack:
  def __init__(self):
    self.top = None

  def push(self, value):
    new_node = Node(value, None)
    new_node.next = self.top
    self.top = new_node

  def pop(self):
    if self.top == None:
      return None

    value = self.top.element
    self.top = self.top.next
    return value

  def peek(self):
    if self.top == None:
      return None

    return self.top.element

  def is_empty(self):
    return self.top == None
```

```python
stack = Stack()
stack.push(3)
stack.push(5)
stack.push(7)
stack.push(9)
```

```python
def string_rev(str):
  stack = Stack()

  for i in range(len(str)):
    stack.push(str[i])

  rev_str = ""

  while not stack.is_empty():
    rev_str += stack.pop()

  return rev_str
```

```python
print(string_rev("DataStrictures"))
```

```python
def evaluate(var1, var2, operator):
  if operator == "+":
    return var1+var2

  if operator == "-":
    return var1-var2

  if operator == "*":
    return var1*var2

  if operator == "/":
    return var1/var2
```

```python
def eval_postfix(expression):
  stack = Stack()
  operators = ['+', '-', '*', '/']

  for i in range(len(expression)):
    if expression[i] in operators:
      var2 = stack.pop()
      var1 = stack.pop()
      result = evaluate(var1, var2, expression[i])
      stack.push(result)

    else:
      stack.push(int(expression[i]))

  return stack.pop()

```

```python
eval_postfix("35+2/")
```
