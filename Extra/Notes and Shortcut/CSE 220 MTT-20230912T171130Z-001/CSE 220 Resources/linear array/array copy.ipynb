```python
#copy array

arr = [10, 20 , 30 , 40 ,50]

def copy_array(source):
  new_array = [None]*len(source)
  # [none, none, none, none, none]
  for i in range(0, len(source), 1):
    new_array[i] = source[i]

  return new_array

print(copy_array(arr))
```

```python
arr = [10, 20 , 30 , 40 ,50]

def reverse_out_of_place(arr):
  rev_arr = [None]*len(arr)

  j = len(arr) - 1
  for i in range(0, len(arr), 1):
    rev_arr[j] = arr[i]  #rev_arr[len(arr)-1-i] = arr[i]
    j = j - 1

  return rev_arr

print(reverse_out_of_place(arr))
```

```python

arr = [10, 20 , 30 , 40 ,50]

def reverse_in_place(arr):
  l = len(arr)
  for i in range(0, l//2, 1 ):
    temp = arr[i]
    arr[i] = arr[l-1-i]
    arr[l-1-i] = temp

  return arr
```

```python
def right_rotate(arr):
  temp = arr[len(arr)-1]
  for i in range(len(arr)-1, 0, -1):
    arr[i] = arr[i-1]

  arr[0] = temp
  return arr
```

```python
import numpy as np
sequence=np.array([10,20,30,40,50,60])
print(right_rotate(sequence))
```

```python
def mergeLineup(pokemon_1, pokemon_2):
  pokemon_1_index = 0
  pokemon_2_index = len(pokemon_2)-1

  while(pokemon_1_index < len(pokemon_1)):
    if pokemon_1[pokemon_1_index] == None:
      pokemon_1[pokemon_1_index] = 0
    if pokemon_2[pokemon_2_index] == None:
      pokemon_2[pokemon_2_index] = 0
    pokemon_1[pokemon_1_index] = pokemon_1[pokemon_1_index] + pokemon_2[pokemon_2_index]
    pokemon_1_index += 1
    pokemon_2_index -= 1

  return pokemon_1
```

```python
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}')
```
