```python
def prime_check(n, divisor):
  if n < 2:
    return False

  if divisor == n:
    return True

  if n % divisor == 0:
    return False

  return prime_check(n, divisor+1)

```

```python
# Python3 program to check whether a number
# is prime or not using recursion

# Function check whether a number
# is prime or not


def isPrime(n, i):

	# Corner cases
	if (n == 0 or n == 1):
		return False

	# Checking Prime
	if (n == i):

		return True

	# Base cases
	if (n % i == 0):
		return False

	i += 1

	return isPrime(n, i)


for n in range(1, 20):
    if (isPrime(n, 2)) == True:
      print(n, end= " ")
    else:
      print(end= "")

```
