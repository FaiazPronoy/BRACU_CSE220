```python
# Problem 14
def sub_array_match(arr, sub, indx):

  if indx+len(sub)-1 >= len(arr):
    return 0


  j = 0
  flag = 1
  for i in range(indx, indx+len(sub)):
    if arr[i]!=sub[j]:
      flag = 0
      break
    j+=1



  return flag + sub_array_match(arr, sub, indx+1)
```

```python
sub_array_match(arr="atata", sub="ata", indx=0)
```

```python
#ques 28
def divide_into_grps(arr, sgrp1, sgrp2, indx):

  if indx==len(arr):

    if sgrp1%10 ==0 and sgrp2%2 ==1:
      return True

    if sgrp1%2 ==1 and sgrp2%10 ==0:
      return True

    return False

  if divide_into_grps(arr, sgrp1+arr[indx], sgrp2, indx+1) or divide_into_grps(arr, sgrp1, sgrp2+arr[indx], indx+1):
    return True

  return False
```

```python
divide_into_grps([5,5,5], 0,0,0)
```

```python
#problem25
def strdist(str, sub, indx, start, end):

  if indx==len(str):
    return (end-start)

  j=0
  for i in range(indx, indx+len(sub)):
    if str[i]!=sub[j]:
      return strdist(str, sub, indx+1, start, end)
    j+=1

  if start > indx:
    start = indx

  if end <  indx+len(sub):
    end = indx+len(sub)

  return strdist(str, sub, indx+len(sub)-1, start, end)
```

```python
strdist("catcowcatcata", "cat", 0, len('catcowcat'), -1)
```

```python
#problem 27
def balance_sum(arr, grp1_sum, grp2_sum, indx):
  if indx==len(arr):
    return grp1_sum == grp2_sum


  if_included_in_grp1 = balance_sum(arr, grp1_sum+arr[indx], grp2_sum, indx+1)
  if_included_in_grp2 = balance_sum(arr, grp1_sum, grp2_sum+arr[indx] , indx+1)

  return if_included_in_grp1 or if_included_in_grp2
```

```python
balance_sum([5, 2,3, 6], 0, 0, 0)
```

```python
#ques 24
def is_nested(str, i , j):

  if i>=j:
    return True

  if str[i] == '(' and str[j] == ')':
    return is_nested(str, i+1, j-1)

  return False
```

```python
is_nested("((()))", i=0, j=len("((()))")-1)
```
